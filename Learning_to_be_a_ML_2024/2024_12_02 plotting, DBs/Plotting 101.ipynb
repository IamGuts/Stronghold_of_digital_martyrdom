## Matplotlib
https://matplotlib.org/

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Simple graph

```python
x = np.linspace(1,25,50)
y = np.sin(x) + np.random.rand()
fig = plt.figure()
plt.plot(x,y)
plt.show()
```

### Size of canvas

```python
fig = plt.figure(figsize=(12,3))
plt.plot(x,y, 'v--', lw=0.75, color='red')
```

### X/Y labels

```python
fig = plt.figure(figsize=(12,3))
plt.plot(x,y, 'v--', lw=0.75, color='red')
plt.xlabel('X-dim', fontsize=12)
plt.ylabel('Y-dim',fontsize=12)
plt.show()
```



### Grid.

```python
fig = plt.figure(figsize=(12,3))
plt.grid(visible=True)
plt.plot(x,y, 'x--', lw=0.75, color='red')
plt.xlabel('X-dim', fontsize=12)
plt.ylabel('Y-dim',fontsize=12)
```

### Legend

```python
fig = plt.figure(figsize=(12,3))
plt.grid(visible=True)
plt.plot(x,y, 'x--', lw=0.75, color='red', label='X-dim vs. Y-dim')
plt.xlabel('X-dim', fontsize=12)
plt.ylabel('Y-dim',fontsize=12)
plt.legend(loc='upper right')
```

```python
z = np.cos(x) + 2 * np.cos(x)
fig = plt.figure(figsize=(12,3))
plt.grid(visible=True)
plt.plot(x,y, 'x--', lw=0.75, color='red', label='X-dim vs. Y-dim')
plt.plot(x,z,color='violet', label = 'X-dim vs. Z-dim')
plt.xlabel('X-dim', fontsize=12)
plt.ylabel('Y-dim',fontsize=12)
plt.legend(loc='upper left', ncols=2)
```

### Xlim/Ylim

```python
fig = plt.figure(figsize=(12,4))
plt.grid(visible=True)
plt.plot(x,y, 'x--', lw=0.75, color='red', label='X-dim vs. Y-dim')
plt.plot(x,z,color='violet', label = 'X-dim vs. Z-dim')
plt.xlim(5, 15)
plt.ylim(-2,2)
plt.xlabel('X-dim', fontsize=12)
plt.ylabel('Y-dim',fontsize=12)
plt.legend(loc='lower left', ncols=2)
```

### Different types of graphs

#### Bar plot

```python
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 12), xticks=np.arange(1, 8),
       ylim=(0, 12), yticks=np.arange(1, 8))

plt.show()
```

#### Stack plot

```python
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])


fig, ax = plt.subplots()

ax.stackplot(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

#### Histogramm

```python
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)


fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

#### Pie plot

```python
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))


fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

#### 3D scatter plot

```python
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

## Seaborn
https://seaborn.pydata.org/

```python
import seaborn as sns
```

### Seaborn dataset

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

### Simple scatterplot

```python
sns.set_style('whitegrid')
```

```python
sns.scatterplot(data=df, x='bill_length_mm', y='flipper_length_mm')
```

### Relplot. Kind - scatter

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter')
```

#### Adding column (col)

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter',col='sex')
plt.show()
```

#### Adding hue by categorical column (species)

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter', hue='species', col='sex')
```

#### Adding hue by number column (body_mass_g)

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter', hue='body_mass_g')
```

#### Using size instead of hue

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter', size='body_mass_g', hue='sex')
```

```python
sns.boxplot(data=df, x='bill_length_mm')
```
